```python
def fix_store(x):
#    return x.split('-')[1].title().strip()
    return x.title()

all_stores['Store'] = all_stores['Store'].apply(fix_store)
```

```python
def fix_price(x):
    return x.split('/')[0]

all_stores['Price'] = all_stores['Price'].apply(fix_price)
```

```python
def fix_package(x):
    if 'PACKAGE' in x:
        return 'Full'
    if 'KIDS' in x:
        return 'Junior'
    if 'SKIS ONLY' in x:
        return 'Skis only'
    if 'BOOT ONLY':
        return 'Boot only'
    
all_stores['Package_Type'] = all_stores['Package'].apply(fix_package)
```

```python
def fix_package_again(x):
     return x.split('DETAILS')[0].title()
```

```python
all_stores['Store Group'] = 'Christy'
all_stores['Category'] = 'Ski'
all_stores['Package'] = all_stores['Package'].apply(fix_package_again)
```

```python
vail_stores1 = ['Lionshead Village Vail', 'Vail And Beaver Creek Delivery', 'Vail', 'Vail Village Bridge Street']
breck_stores1 = ['Breck Village', 'Breck Lone Star', 'Main St. Station Breck', 'Breckenridge Delivery', 'Grand Timber Breck', 'Pine Ridge Breck', 'At The Ranahan Breck', 'Frisco']
park_city_stores1 = ['Deer Valley', 'Snowbird', 'Olympus Hills', 'Kimball Junction', 'Park City', 'Park City And Deer Valley Delivery']
front_range_stores1 = ['Arvada', 'Boulder', 'Cherry Creek', 'Denver West', 'Fort Collins', 'Park Meadows', 'Littleton Event Center', 'Colorado Springs']
crested_butte_stores1 = ['christy-sports-treasury-center.html', 'crested-butte.html', 'the-alpineer-crested-butte.html']
aspen_stores1 = ['Snowmass Village', 'Woodrun Place Condos Snowmass', 'Stonebridge Condos', 'Aspen And Snowmass Delivery']
keystone_stores1 = ['Snake River Keystone']
beaver_creek_stores1 = ['Avon', 'Beaver Creek']


def get_resort(x):
    x = x.split('-')[1].title().strip()
    if x in vail_stores1:
        return 'Vail'
    if x in breck_stores1:
        return 'Breckenridge'
    if x in park_city_stores1:
        return 'Park City'
    if x in front_range_stores1:
        return 'Front Range'
    if x in crested_butte_stores1:
        return 'Crested Butte'
    if x in aspen_stores1:
        return 'Aspen'
    if x in keystone_stores1:
        return 'Keystone'
    if x in beaver_creek_stores1:
        return 'Beaver Creek'

all_stores['Resort'] = all_stores['Store'].apply(get_resort)



```

```python
def get_product_category(x):
    if 'Kids' in x:
        return 'Junior'
    if 'Demo' in x:
        return 'Demo'
    if 'Sport' in x:
        return 'Sport'
    if 'Premium' in x:
        return 'Premium'
    else:
        return 'Boots'
    
```

```python
all_stores['Product Category'] = all_stores['Package'].apply(get_product_category)
```
